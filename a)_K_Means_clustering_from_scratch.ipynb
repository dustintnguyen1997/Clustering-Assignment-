```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("coin_Dogecoin.csv")
df.head()
```

```python
plt.scatter(df.SNo,df['Marketcap'])
plt.xlabel('SNo')
plt.ylabel('Marketcap')
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['SNo','Marketcap']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
km.cluster_centers_
```

```python
plt.plot(km.cluster_centers_)
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.SNo,df1['Marketcap'],color='green')
plt.scatter(df2.SNo,df2['Marketcap'],color='red')
plt.scatter(df3.SNo,df3['Marketcap'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('SNo')
plt.ylabel('Marketcap')
plt.legend()
```

```python
scaler = MinMaxScaler()

scaler.fit(df[['Marketcap']])
df['Marketcap'] = scaler.transform(df[['Marketcap']])

scaler.fit(df[['SNo']])
df['SNo'] = scaler.transform(df[['SNo']])
```

```python
df.head()
```

```python
plt.scatter(df.SNo,df['Marketcap'])
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['SNo','Marketcap']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.SNo,df1['Marketcap'],color='green')
plt.scatter(df2.SNo,df2['Marketcap'],color='red')
plt.scatter(df3.SNo,df3['Marketcap'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.legend()
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['SNo','Marketcap']])
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
